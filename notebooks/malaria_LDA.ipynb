```python
import numpy as np
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import GridSearchCV
from scipy.stats import entropy
```

```python
data = pd.read_csv("malaria/pb10xIDC_counts.csv",header=0,index_col=0)
```

```python
data
```

```python
metadata = pd.read_csv("malaria/pb10xIDC_pheno.csv",header=0,index_col=0)
```

```python
metadata = metadata.sort_values(['clock_pseudotime'],ascending=True)
```

```python
lda = LatentDirichletAllocation(n_components=20,doc_topic_prior=0.01,topic_word_prior=0.05,learning_method="online",
                               random_state=42)
```

```python
output = lda.fit_transform(data.T)
```

```python
data
```

```python
output.shape
```

```python
word_topic = lda.components_/lda.components_.sum(axis=1)[:,np.newaxis]
```

```python
topicsDocs = pd.DataFrame(output, index=data.columns)
```

```python
topicsDocs = topicsDocs.reindex(index=metadata.index)
```
